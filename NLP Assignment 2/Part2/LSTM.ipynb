```python
# Specify the path to your JSON file
json_file_path = '../NER_training_data.json' 

# Read the JSON file
with open(json_file_path) as f:
    data = json.load(f)

data

```

```python
max_len=0
for key,value in data.items():
    if len(value['text'].split(" "))>max_len:
        max_len=len(value['text'].split(" "))

print(max_len)
```

```python
data_for_training = []
vocab = []
for key,value in data.items():
    l=[]
    s=value['text'].split(" ")
    for j in range(len(s)):
        k=[]
        k.append(s[j])
        vocab.append(s[j])
        k.append(value['labels'][j])
        l.append(k)
    if(len(l)<max_len):
        u=max_len-len(l)
        for n in range(u):
            k=[]
            k.append("")
            k.append("O")
            l.append(k)
    data_for_training.append(l)
# vocab = list(set(vocab))
# data_for_training
# vocab
```

```python
model = KeyedVectors.load_word2vec_format('./GoogleNews-vectors-negative300.bin', binary=True)
vocab_weight = {}
word='hello'

for i in vocab:
    if i in model:
        vocab_weight[i] = model.get_vector(i)
    else:
        vocab_weight[i] = [0]*300
```
